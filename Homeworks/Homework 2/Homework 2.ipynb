### Calculations for Question 2

Question 2a: Condition Number

```python
A = np.matrix([[1/2, 1/2], [1/2*(1 + 10**-10), 1/2*(1 - 10**-1)]])
A_inverse = np.matrix([[1 - 10**10, 10**10], [1 + 10**10, -10**10]])

K = la.norm(A_inverse)*la.norm(A)

K
```

Question 2c: Relative error

```python
x = np.matrix([[1], [1]])
delta_x = np.matrix([[10**5 + 10**-5], [-10**5 + 10**-5]])
error = la.norm(delta_x)/la.norm(x)
error
```

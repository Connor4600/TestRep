```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
x = np.linspace(0, 5, 1000)

fex = lambda x: np.sin(x)
T = lambda x: x - (1/6)*x**3 + (1/120)*x**5
P33 = lambda x: (x - (7/60)*x**3)/(1 + (1/20)*x**2)
P24 = lambda x: x/(1 + (1/6)*x**2 - (7/360)*x**4)
P42 = lambda x: (x - (7/60)*x**3)/(1 + (1/20)*x**2)
```

```python
errT = abs(fex(x) - T(x))
errP33 = abs(fex(x) - P33(x))
errP24 = abs(fex(x) - P24(x))
errP42 = abs(fex(x) - P42(x))
```

```python
plt.plot(x, fex(x), label = 'Sin(x)')
plt.plot(x, T(x), label = 'Taylor')
plt.plot(x, P33(x), label = 'P33')
plt.plot(x, P24(x), label = 'P24')
plt.plot(x, P42(x), label = 'P42')
plt.ylim(-3,3)
plt.legend()
plt.show()
```

```python
plt.plot(x, fex(x)-fex(x))
plt.plot(x, errT, label = 'Error Taylor')
plt.plot(x, errP33, label = 'Error P33')
plt.plot(x, errP24, label = 'Error P24')
plt.plot(x, errP42, label = 'Error P42')
plt.ylim(-3,3)
plt.legend()
plt.show()
```

```python
def Comp_Trap(a, b, n):
    h = (b - a)/n
    x = np.linspace(a, b, n + 1)
    f = lambda x: 1/(1 + x**2)
    
    func = 0
    for i in range(1, n):
        func += f(x[i])
        
    return h/2 * (f(a) + 2*func + f(b))

Comp_Trap(-5, 5, 1291)
    

    
```

```python
def Comp_Simp(a, b, n):
    h = (b - a)/n
    x = np.linspace(a, b, n + 1)
    f = lambda x: 1/(1 + x**2)
    n1 = int(n/2 - 1)
    n2 = int(n/2)
    x1 = x[::2]
    x2 = x[1::2]

    func1 = 0
    func2 = 0
    for i in range(n1):
        func1 += f(x1[i])
    for i in range(n2):
        func2 += f(x2[i])
        
    return h/3 * (f(a) + 2*func1 + 4*func2 + f(b))

Comp_Simp(-5, 5, 108)
```

```python
fex = lambda x: 1/(1 + x**2)
Integral1 = quad(fex, -5, 5, full_output = 1, epsabs = 10**-4)
Integral2 = quad(fex, -5, 5, full_output = 1, epsabs = 10**-6)
```

```python
err_Comp_Trap1 = abs(quad(fex, -5, 5, epsabs = 10**-6)[0] - Comp_Trap(-5, 5, 1291))
err_Comp_Trap2 = abs(quad(fex, -5, 5, epsabs = 10**-4)[0] - Comp_Trap(-5, 5, 1291))
err_Comp_Simp1 = abs(quad(fex, -5, 5, epsabs = 10**-6)[0] - Comp_Simp(-5, 5, 108))
err_Comp_Simp2 = abs(quad(fex, -5, 5, epsabs = 10**-4)[0] - Comp_Simp(-5, 5, 108))
print('Comp Trap Error for tol = 1e-7: ', err_Comp_Trap1)
print('Comp Trap Error for tol = 1e-5: ', err_Comp_Trap2)
print('Comp Simp Error for tol = 1e-7: ', err_Comp_Simp1)
print('Comp Simp Error for tol = 1e-5: ', err_Comp_Simp2)
```
